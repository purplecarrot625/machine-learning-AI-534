```python
a = np.array([4,2])
plt.plot((0, a[0]), (0,a[1]))
```

```python
plt.quiver(0,0,a[0],a[1]) #
```

```python
plt.quiver(0,0,a[0],a[1], angles='xy', scale_units='xy', scale=1, color='c') #
```

```python
def arrow(start, end, color='b'):
    x,y = start
    x1,y1 = end
    plt.quiver(x,y,x1,y1, angles='xy', scale_units='xy', scale=1, color=color)
```

```python
arrow((0,0),(-2,2))
```

```python
arrow((0,0), np.array([-1,11]))
```
